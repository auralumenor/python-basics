```python
# Program to print Fibonacci series
a=int(input("Enter a Number: "))
def fib_ite(a):
 fib = [0,1]
 for i in range (2,a):
   next = fib[i-1] + fib[i-2]
   fib.append(next)
 return fib
print (fib_ite(a))
```
