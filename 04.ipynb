```python
import random
n=10
numbers=[random.randint(1,100) for _ in range(n)]
print("Numbers: ",numbers)
avg=sum(numbers)/len(numbers)
print("Average: ",avg)
count=0
for num in numbers:
  if num>avg:
    count+=1
print("Count of numbers greater than average: ",count)
```

```python
words=[]
n1=int(input("Enter the number of words: "))
for i in range (n1):
  sub=input("Enter the words: ")
  words.append(sub)
print(words)
shortest=min(words,key=len)
print("The shortest word is: ",shortest)
```
