```python
book=[]
years=[]
price=[]
n=int(input("Enter the number of books: "))
for i in range(n):
  name=input("Enter the name of the book: ")
  year=int(input("Enter the year it published: "))
  cost=int(input("Enter the Cost of the Book: "))
  book.append(name)
  years.append(year)
  price.append(cost)
print("Books: ",book)
print("Year when it published: ",years)
print("Cost of the Book: ",price)
for i in range(len(years)-1, -1, -1):
  if years[i]<2020:
    del book[i]
    del years[i]
    del price[i]
print("Books: ",book)
print("Year when it published: ",years)
print("Cost of the Book: ",price)
```
