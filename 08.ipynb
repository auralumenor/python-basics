```python
#program to find factorial of number using function

def fac(a):
  r=1
  while(a>0):
    if (a!=0):
      r*=a
      a-=1
  return r
a=int(input("Enter a number: "))
print("The Factorial of given number is: ",fac(a))
```

```python
#Program to find whether the given number is prime  or not

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

num = int(input("Enter a number: "))
print("Prime?", is_prime(num))
```

```python
#Program to find Fibonacci series
def fibonacci(n):
    a, b = 0, 1
    series = []
    for _ in range(n):
        series.append(a)
        a, b = b, a + b
    return series

terms = int(input("Enter number of terms: "))
print("Fibonacci series:", fibonacci(terms))
```

```python
#Program to swap two numbers
def swap(a, b):
    return b, a

x = int(input("Enter first number: "))
y = int(input("Enter second number: "))
x, y = swap(x, y)
print("After swapping: x =", x, ", y =", y)
```

```python
#Program to find amicable number
def sum_of_divisors(n):
    return sum(i for i in range(1, n) if n % i == 0)

def are_amicable(x, y):
    return sum_of_divisors(x) == y and sum_of_divisors(y) == x

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print("Amicable?", are_amicable(a, b))
```

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

def is_strong(n):
    return sum(factorial(int(d)) for d in str(n)) == n

num = int(input("Enter a number: "))
print("Strong number?", is_strong(num))
```
