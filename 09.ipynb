```python
#program to find factorial of a number
def factorial(n: int) -> int:
    if n < 0:
        return -1
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)
if __name__ == "__main__":
    try:
        num = int(input("Enter a non-negative integer: "))
        print(f"{num}! = {factorial(num)}")
    except ValueError as e:
        print("Error:", e)
```

```python
#program to find fibonacci series
def fib(n):
    if n < 0:
        return -1
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fib(n-1) + fib(n-2)
if __name__ == "__main__":
    for i in range(10):
        print(i, fib(i))
```

```python
#program to find sum of n natural numbers
def sumn(n: int) -> int:
    if n <= 0:
        return 0
    return n + sumn(n - 1)
if __name__ == "__main__":
    N = int(input("Enter N: "))
    print(f"Sum of first {N} natural numbers is:", sumn(N))
```

```python
#program to find sum of the numbers in a list
def lsum(nums):
    if not nums:
        return 0
    return nums[0] + lsum(nums[1:])
lst = [1, 2, 3, 4, 5]
print(lsum(lst))
```

```python
#program to find maximum number in a list
def max(lst):
    if not lst:
        return -1
    if len(lst) == 1:
        return lst[0]
    first = lst[0]
    maxr = max(lst[1:])
    return first if first > maxr else maxr
print(max([3, 1, 4, 1, 5, 9]))
print(max([-2, -7, -1, -5]))
```
