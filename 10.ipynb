```python
adder = lambda x, y: x + y
print(adder(5, 3))
multiplier = lambda x, y: x * y
print(multiplier(6, 8))
```

```python
l = [1,2,3,4,5,6,7,8,9]
odd = list(filter(lambda x: x % 2 != 0, l))
print(odd)
```

```python
z=[1,2,3,4,5,6,7,8,9]
doub=list(map(lambda x: x*2,z))
print(doub)
```

```python
tl=[('a',34),('b',98),('c',56),('d',74),('z',94)]
sl=sorted(tl, key=lambda x: x[1])
print(sl)
```

```python
import pandas as pd
students = {
    "ansh": 82,
    "arun": 74,
    "ajay": 91,
    "amit": 68,
    "suraj": 77
}
s = pd.Series(students)
above_75 = s[s > 75]
print("All percentages:\n", s)
print("\nAbove 75%:\n", above_75)
```
